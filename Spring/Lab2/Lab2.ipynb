- точечная диаграмма (мало данных)

````python
import numpy as np
solved = [8, 8, 8, 8, 8, 8, 8, 8, 8, 8, 8, 8, 8, 7, 7, 7, 7, 7, 7, 7, 7, 7, 7,
          7, 6, 6, 6, 6, 6, 6, 6, 6, 6, 6, 6, 6, 6, 6, 6, 5, 5, 5, 5, 5, 5, 5,
          4, 4, 4, 4, 4, 4, 4, 4, 3, 3, 3, 3, 3, 3, 3, 3, 3, 3, 2, 2, 2, 1, 1,
          1, 1, 0, 0, 0, 0, 0]
np_solved = np.array(solved) # преобразование к np.array
values, counts = np.unique(np_solved, return_counts=True) #создание пары массивов:
#                                          уникальные элементы и их количество
n = len(values) # количество уникальных элементов
purple = 'purple'
for i in range(n): # для каждого значения
    for j in range(counts[i]): # рисуем соответсвующее количество точек
        plt.scatter(values[i], j, color=purple) # красим в единый цвет
plt.title('Количество студентов 2-го курса,\nрешивших определённое число задач')
plt.xlabel('Число задач')
plt.ylabel('Количество студентов')
plt.show()
````

- точечный график (две переменные)

````python
plt.figure(figsize=(15, 5))

# случай 1
x1 = [1, 2, 3, 4, 5]
y1 = [0.99, 0.49, 0.35, 0.253, 0.18]
sp1 = plt.subplot(131)
sp1.scatter(x1, y1) # рисуются  отдельные точки
plt.title('Малое количество точек')
plt.xlabel('Независимая переменная')
plt.ylabel('Отклик')

# случай 2
x2 = np.random.rand(100) # 100 случайных чисел от 0 до 1
y2 = np.random.rand(100) # tещё 100 случайных чисел от 0 до 1
sp1 = plt.subplot(132)
sp1.scatter(x2, y2) # рисуются  отдельные точки
plt.title('Переменные не связаны между собой')
plt.xlabel('Независимая переменная')
plt.ylabel('Отклик')

# случай 3
data = 10 * np.random.rand(100) # 100 случайных чисел от 0 до 10
x3 = np.sort(data)              # отсортировали
epsilon3 = np.random.rand(100)  # 100 случайных чисел от 0 до 1
y3 = (x3-5)**2 / 10 + epsilon3 - 0.5
sp1 = plt.subplot(133)
sp1.plot(x3, y3) # рисуется линия
plt.title('Переменная')
plt.xlabel('Независимая переменная')
plt.ylabel('Отклик')
plt.show()
````

## Корреляция

**Корреляция** — статистическая взаимосвязь двух или более случайных величин (либо величин, которые можно с некоторой допустимой степенью точности считать таковыми). При этом изменения значений одной или нескольких из этих величин сопутствуют систематическому изменению значений другой или других величин.
